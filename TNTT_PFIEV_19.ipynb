```python
#Trò chơi 8 số
import random as rand

def ZeroPos(O):
  for i in range(n):
    for j in range(n):
      if O[i][j]==0:
        return i,j

def Up(O):
  i,j = ZeroPos(O)
  if i<n-1:
    N = [O[_][:] for _ in range(n)]
    N[i][j] = N[i+1][j]
    N[i+1][j] = 0
    return N
  else:
    return None

def Down(O):
  i,j = ZeroPos(O)
  if i>0:
    N = [O[_][:] for _ in range(n)]
    N[i][j] = N[i-1][j]
    N[i-1][j] = 0
    return N
  else:
    return None

def Left(O):
  i,j = ZeroPos(O)
  if j<n-1:
    N = [O[_][:] for _ in range(n)]
    N[i][j] = N[i][j+1]
    N[i][j+1] = 0
    return N
  else:
    return None

def Right(O):
  i,j = ZeroPos(O)
  if j>0:
    N = [O[_][:] for _ in range(n)]
    N[i][j] = N[i][j-1]
    N[i][j-1] = 0
    return N
  else:
    return None

def getChildren(O):
  ans = []
  if Up(O)!=None:
    ans.append(Up(O))
  if Down(O)!=None:
    ans.append(Down(O))
  if Left(O)!=None:
    ans.append(Left(O))
  if Right(O)!=None:
    ans.append(Right(O))
  return ans

n = 3
Goal = [[(i*n+j+1)%(n*n) for j in range(n)]for i in range(n)]
Start = [Goal[i][:] for i in range(n)]
for _ in range(1000):
  r = rand.randint(0,3)
  if r == 0:
    Next = Up(Start)
  elif r == 1:
    Next = Down(Start)
  elif r == 2:
    Next = Left(Start)
  elif r ==3:
    Next = Right(Start)
  if Next != None:
    Start = Next
```

```python
# A* algorithm
from queue import PriorityQueue

def printAll(O):
  Banthan = O[2]
  ToTien = O[3]
  if ToTien!=None:
    printAll(ToTien)
  for i in range(n):
    for j in range(n):
      print(Banthan[i][j],end=" ")
    print()
  print("----------")

def h(O):
  c = 0
  for i in range(n):
    for j in range(n):
      if O[i][j] !=0 and O[i][j] != Goal[i][j]:
        c += 1
  return c

def f(O,g):
  return g + h(O)

Open = PriorityQueue()
Closed = set()
OK = False
# 1. Cho đỉnh xuất phát vào open. 
Open.put((f(Start,0),0,Start,None))
Closed.add(str(Start))
count = 0
# 2 - 6: Nếu open rỗng thì tìm kiếm thất bại, kết thúc việc tìm kiếm.
while not Open.empty():
  count += 1
  # 3. Lấy đỉnh đầu trong open ra và gọi đó là ʘ. Cho ʘ vào closed
  O = Open.get()
  # 4. Nếu ʘ là đỉnh đích thì tìm kiếm thành công, kết thúc việc tìm kiếm.
  if O[2]==Goal:
    OK = True
    break
  # 5. Tìm tất cả các đỉnh con của ʘ, không thuộc open và closed cho vào cuối của open
  children = getChildren(O[2])
  for child in children:
    if str(child) not in Closed:
      #Open.insert(rand.randint(0,len(Open)),(child,O))
      
      Open.put((f(child,O[1]+1),O[1]+1,child,O)) # tìm kiếm theo chiều rộng
      # Open.insert(0,(child,O)) # tìm kiếm theo chiều sâu
      Closed.add(str(child))

if OK:
  print("Tìm kiếm thành công")
  printAll(O)
else:
  print("Tìm kiếm thất bại")
print(count)
```

```python
data = [[0,4],[1,5],[4,0],[4,1]]
A,B,C,D = data

def TrongTam(X):
  dim = len(X[0])
  res = [0]*dim
  for x in X:
    for i in range(dim):
      res[i] += x[i]/len(X)
  return res

def distance(x,y):
  dim = len(x)
  res = 0
  for i in range(dim):
    res += (x[i]-y[i])**2
  return res

def SumDis2TT(X):
  TT = TrongTam(X)
  res = 0
  for x in X:
    res += distance(x,TT)
  return res

print("ABC - D:", SumDis2TT([A,B,C])+SumDis2TT([D]))
print("ABD - C:", SumDis2TT([A,B,D])+SumDis2TT([C]))
print("ACD - B:", SumDis2TT([A,C,D])+SumDis2TT([B]))
print("BCD - A:", SumDis2TT([B,C,D])+SumDis2TT([A]))
print("AB - CD:", SumDis2TT([A,B])+SumDis2TT([C,D]))
print("AC - BD:", SumDis2TT([A,C])+SumDis2TT([B,D]))
print("AD - BC:", SumDis2TT([A,D])+SumDis2TT([B,C]))
```

```python
import random as rd
data = [[0,4],[4,0],[1,5],[4,1]]

def distance(x,y):
  dim = len(x)
  res = 0
  for i in range(dim):
    res += (x[i]-y[i])**2
  return res

def argmin(x,C):
  res = 0
  res_min = distance(x,C[res])
  for i in range(len(C)):
    d = distance(x,C[i])
    if res_min > d:
      res = i
      res_min = d
  return res

def kmean(data,k):
  # Buoc 0
  l = [rd.randint(0,k-1) for _ in data]

  while True:
    # Buoc 1
    C = [None]*k
    num = [0]*k
    for i in range(len(data)):
      if C[l[i]] == None:
        C[l[i]] = [0]*len(data[i])
      for j in range(len(data[i])):
        C[l[i]][j] += data[i][j]
      num[l[i]] += 1
    
    for i in range(k):
      if num[i] !=0:
        for j in range(len(C[i])):
          C[i][j] /= num[i]
      else:
        C[i] = data[rd.randint(0,len(data)-1)][:]
      
    KiemTra = True
    # Buoc 2
    for i in range(len(data)):
      l_new = argmin(data[i],C)
      if l[i] != l_new:
        KiemTra = False
      l[i] = l_new
    if KiemTra:
      break
  return C,l

print(kmean(data,2))
```
